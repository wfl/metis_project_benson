```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import pickle
```

```python
sns.set(style='white', font='sans-serif',font_scale=3)
sns.despine()
#plt.tight_layout()
```

## Demographics data

```python
with open('plot_f.pickle','rb') as read_file:
    plot_f = pickle.load(read_file)  
```

```python
plt.figure(figsize=(40,10))
clrs = ['red' if (x > 70) else 'mediumblue' for x in plot_f['working adult F']]
fig1 = sns.barplot(x ='District',y ='working adult F',data=plot_f,palette=clrs)
fig1.set_xlabel("Districts",fontsize=50)
fig1.set_ylabel("% Women in Work Force",fontsize=50)
fig1.set(xticks=[])
fig1
plt.savefig('districts_women.png',dpi=300,transparent=True)
```

```python
plt.figure(figsize=(40,10))
fig2 = sns.barplot(x ='District',y ='working adult F',data=plot_f,palette='bone')
fig2.set_xlabel("Districts",fontsize=50)
fig2.set_ylabel("% Women in Work Force",fontsize=50)
fig2.set(xticks=[])
fig2
plt.savefig('districts_women.png',dpi=300,transparent=True)
```

```python
with open('plot_industry','rb') as read_file:
    plot_industry = pickle.load(read_file) 
```

```python
plt.figure(figsize=(40,10))
clrs = ['red' if (x > 70) else 'mediumblue' for x in plot_industry['ppl in sci bus arts']]
fig3 = sns.barplot(x ='District',y ='ppl in sci bus arts',data=plot_industry,palette=clrs)
fig3.set_xlabel("Districts",fontsize=50)
fig3.set_ylabel("% Employed in Selected Industries",fontsize=50)
fig3.set(xticks=[])
fig3
plt.savefig('districts_industry.png',dpi=300,transparent=True)
```
